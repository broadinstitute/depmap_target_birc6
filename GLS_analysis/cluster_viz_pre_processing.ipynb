## Individual cluster visualization

```python
selected_correlations.index.name = 'gene'
selected_correlations.head()
```

```python
cluster_file = pd.read_csv('data/cluster_2000.csv', index_col=0)
cluster_file.head()
```

```python
from pathlib import Path
Path("data/cluster_level/").mkdir(parents=True, exist_ok=True)
```

```python
for rank,cluster_list in enumerate(cluster_file.community_string):
    cluster_list = cluster_list.split(',')
    cluster = selected_correlations.loc[cluster_list,cluster_list]
    cluster = cluster.reset_index('gene').melt(id_vars='gene', var_name='to', value_name='p_value')
    cluster = cluster.loc[cluster.p_value!=0]
    
    cluster.to_csv('data/cluster_level/cluster_{}.csv'.format(rank))
```
